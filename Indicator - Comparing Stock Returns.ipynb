# Využití indexů v grafech

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
#GSPC - S&P 500
#IXIC - NASDAQ
#GDAXI - German DAX
#FTSE - London FTSE
#DJI - Dow Jones Index
indexy = ["^GSPC", "^IXIC", "^GDAXI", "^FTSE", "^DJI"]
ind_data = pd.DataFrame()

for t in indexy:
    ind_data[t] = wb.DataReader(t, data_source = "yahoo", start="1997-1-1")["Adj Close"]
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
annual_ind_returns = ind_returns.mean() * 250
```

## Srovnání Applu s Indexy S&P 500, DOW, NASDAQ
Kumulativní výnosy

```python
firmy = ["ADBE", "^DJI", "^GSPC", "^IXIC"]
data2 = pd.DataFrame()
for f in firmy:
    data2[f] = wb.DataReader(f, data_source = "yahoo", start = "2007-1-1")["Adj Close"]
    
(data2 / data2.iloc[0] * 100).plot(figsize = (15,6))
plt.show()
```
