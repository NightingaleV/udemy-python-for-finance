## Cost of Goods Sold
Naklady na vyrobu

```python
# Naklady na vyrobu
# 0.6 = COGS Mean
# 0.1 = COGS STD
COGS = - (rev * np.random.normal(0.6, 0.1))
```

```python
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

## Gross Profit
